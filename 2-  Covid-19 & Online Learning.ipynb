# 4- Clean the data frame

    1- Drop records where there are most nan values
    2- replace nan values
    3- change "District_ID" & "Product_ID" columns data types to string.
    3- change time column data type into datetime

#### 4.1- Drop records where there are 20% or more nan values

```python
perc = 20
min_count = int(((100-perc)/100)*final_Covid19_df.shape[1]+1)
final_Covid19_df = final_Covid19_df.dropna(axis=0, thresh=min_count)

final_Covid19_df.isnull().sum()
```

####  4.2-Replace nan values

```python
final_Covid19_df['product category2']= final_Covid19_df['product category2'].fillna('Missing Info')
final_Covid19_df['product category3']= final_Covid19_df['product category3'].fillna('Missing Info')

final_Covid19_df['Sector2']= final_Covid19_df['Sector2'].fillna('Missing Info')
final_Covid19_df['Sector3']= final_Covid19_df['Sector3'].fillna('Missing Info')

final_Covid19_df['Percentage_eligible_free']= final_Covid19_df['Percentage_eligible_free'].fillna('40%-60%')
final_Covid19_df['Residential_connection']= final_Covid19_df['Residential_connection'].fillna('18%-100%')
final_Covid19_df['Per-pupil_expenditure']= final_Covid19_df['Per-pupil_expenditure'].fillna('14000-16000')

final_Covid19_df.isnull().sum()
```

####  4.3-change "District_ID" & "Product_ID" columns data types to string

```python
final_Covid19_df['District_ID'] = final_Covid19_df['District_ID'].astype('str')
final_Covid19_df['Product_ID'] = final_Covid19_df['Product_ID'].astype('str')

final_Covid19_df.info()
```

#### 4.4-change time column data type into datetime

```python
final_Covid19_df['Time'] = pd.to_datetime(final_Covid19_df['Time'])

final_Covid19_df.info()
```

### 6.1-per-pupil_expenditure by state

```python
#create a new data frame with the state and its corresponding per-pupil_expenditures

exp_by_state= pd.crosstab(index= final_Covid19_df['State'], columns= final_Covid19_df['Per-pupil_expenditure'])
exp_by_state
```

```python
#figure out how much does each state spend in total

state_sum = exp_by_state.sum(axis=1 , numeric_only= True)
state_sum.sort_values(ascending=False)
```

###### The states that spend the most expenditure per pupil in total are Connecticut, Utah ,Illinois, and Massachusetts . The least spending state is North Dakota with a total of 1073

```python
%matplotlib notebook

exp_by_state.plot(kind='bar', figsize=(14, 10), width=1.5)
plt.title('per-pupil_expenditure by state', fontsize=12, fontweight='bold')
plt.title('Per-pupil_Expenditure by state', fontsize=11, fontweight='bold')
plt.xlabel('Per-pupil_Expenditure', fontweight='bold', fontsize=10)
plt.show()
```

### 6.2- per-pupil_expenditure by locale

```python
#create a new data frame with the locale and its corresponding per-pupil_expenditure

exp_by_locale = pd.crosstab(index= final_Covid19_df['Locale'], columns= final_Covid19_df['Per-pupil_expenditure'])
exp_by_locale
```

```python
#figure out how much does the government pay for each locale

locale_sum = exp_by_locale.sum(axis=1 , numeric_only= True)
locale_sum.sort_values(ascending=False)
```
